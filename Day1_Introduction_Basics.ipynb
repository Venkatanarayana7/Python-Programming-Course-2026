**Now Let's Talk About Variables  This Is "Storing"**
Remember I said the first superpoewr is Store? This is it.

Imagine a variable like a box. You give the box a name, and you put something inside it. Later, whenever you say the box's name, Python opens it and hands you whatever is inside. 

````python
name = "Alex"
age = 25
````

Here, **name** is a box. Inside that box, you stored the word **"Alex"**. The box called **age** holds the number **25**.

Now watch this:

````python
print(name)
print(age)
````

Python opens those boxes and gives you what's inside. Simple, Clean.Logical.

Now I want you to feel something important here. Notice how **"Alex"** has quotation marks but **25** does not. Why? Because Python needs to know what kind of thing you're storing. Words need quotes. Numbers don't. This is called da **data type** and we'll go deep on this , but right now just feel the logic of it. It makes sense, right? If you write **25** without quotes, Python knows it's a number it can do maht with. If you write **"25"** with quotes, Python treats it like text , it won't add it to another number. 

Try this yourself: 

````python
x = 10
y = 5
print(x + y)
````

You'll get **15**. Now try:



````python
x = "10"
y = "5"
print(x + y)
````

You'll get **105**. Not fifteen. Because now Python thinks you're joining two pieces of text, not adding two numbers. This single concept -data types- is something beginners spend weeks being confused about. You just understood it in two minutes because we went why before what.

*****The Data Types You Must Know Cold*****

There are four basic types you'll use every single day:

Strings -- anything that is text.Always in quotes. **"Hello"**,**"Python"**,**"123 Main Street"**

Integers -- whole numbers. No quotes.**1**,**42**,**1000**

Floats -- decimal numbers. **3.14**, **99.99**, **0.5**

Booleans -- and this one is beautiful. Just two values: **True** *or* **False**. That's it. No quotes. Capital T, capital F. This is how computers make decisions,  everything eventually becomes True or Flase.

```python
is_raining = True
is_sunny = False
temeperature = 98.6
city = "New York"
population = 800000
```

Feel how natural that reads? Python reads almost like English.That's intentional. Guido van Rossum,the person who created Python, wanted it to feel human.

***One More Powerful Thing Before We Close Day 1***

There's something called an **f-string** and once you see it, you'll never want to go back.


````python
name = "David"
age = 22
print(f"My name is {name} and I am {age} years old.")
````

See those curly braces **{}**? You're literally embedding a variable right inside a sentence. The **f** before the quote tell Python,  *"hey,treat this strig as a template and fill in the blanks."* This is how real Python developers write output. Clean,Readable, powerful.
